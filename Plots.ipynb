# Zeros of rescaled Hermite polynomials for $n=40$

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc
from matplotlib.offsetbox import AnchoredText

rc('font',**{'family':'sans-serif','sans-serif':['Helvetica']})
## for Palatino and other serif fonts use:
#rc('font',**{'family':'serif','serif':['Palatino']})
rc('text', usetex=True)
plt.rcParams.update({'font.size': 20})


# Specify the file path
file_path = 'partial-Hermite40.txt'

# Load the data from the file into a NumPy array
data_array = np.loadtxt(file_path, delimiter='\t')
data_array.shape

# Specify the file path
#file_path = 'listas-C-ceros.txt'

# Load the data from the file into a NumPy array
data_array_ceros = np.loadtxt(file_path, delimiter='\t')
data_array_ceros

# Plot Hermite
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import numpy as np

# Create a figure and a subplot
fig, ax = plt.subplots()
ax.grid(True, which='both', linestyle='--', linewidth=0.5, color='gray', alpha=0.5)

def plot_seq2(x, y,a,b):
    
    ## Create a line with gradient transparency
    #for i in range(len(x) - 1):
    #    ax.plot(x[i:i+2], y[i:i+2], linestyle='-', color=(a, b, 1, i / len(x)), linewidth=1.5)
    ax.plot(x, y, linestyle='-', color=(a, b, 1, 0.6), linewidth=2)

    # Highlight the last point with a solid circle
    ax.plot(x[-1], y[-1], 'o', color='black', markersize=3.5)

tmax = 2001
for i in range(0,20):
    plot_seq2(data_array[:tmax,4*i].tolist(),data_array[:tmax,4*i+1].tolist(),0.7,0.2)
    plot_seq2(data_array[:tmax,4*i+2].tolist(),data_array[:tmax,4*i+3].tolist(),0.7,0.2)

ax.set_xlim([-45, 150])
ax.set_ylim([-70, 70])
#ax.plot(data_array_ceros[:,0].tolist(),data_array_ceros[:,1].tolist(),'o', color='black',alpha=0.7, markersize=6)

# Define the function |z e^(1-z)| - 1
def szego_function(z,alpha):
    
    return np.abs(z/alpha * np.exp(1 - z/alpha)) - 1

def plot_zego(alpha):
    # Create a grid of complex numbers
    re = np.linspace(-40, 135, 1000)
    im = np.linspace(-60, 60, 1000)
    re, im = np.meshgrid(re, im)
    z = re + 1j * im

    # Compute the function on the grid
    szego_values = szego_function(z,alpha)

    ax.contour(re, im, szego_values, levels=[0], colors='blue', linewidths=2)

plot_zego(135)


plt.savefig('Hermite-N-40.pdf')
```

# Zeros of rescaled Charlier polynomials for $n=40$

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import rc
from matplotlib.offsetbox import AnchoredText
rc('font',**{'family':'sans-serif','sans-serif':['Helvetica']})
## for Palatino and other serif fonts use:
#rc('font',**{'family':'serif','serif':['Palatino']})
rc('text', usetex=True)
plt.rcParams.update({'font.size': 20})

# Specify the file path
file_path = 'partial-Charlier40.txt'

# Load the data from the file into a NumPy array
data_array = np.loadtxt(file_path, delimiter='\t')
data_array.shape

# Create a figure and a subplot
fig, ax = plt.subplots()
ax.grid(True, which='both', linestyle='--', linewidth=0.5, color='gray', alpha=0.5)

def plot_seq2(x, y,a,b):
    
    ## Create a line with gradient transparency
    #for i in range(len(x) - 1):
    #    ax.plot(x[i:i+2], y[i:i+2], linestyle='-', color=(a, b, 1, i / len(x)), linewidth=1.5)
    ax.plot(x, y, linestyle='-', color=(a, b, 1, 0.6), linewidth=2)

    # Highlight the last point with a solid circle
    ax.plot(x[-1], y[-1], 'o', color='black', markersize=3.5)

tmax = 1000
for i in range(0,20):
    plot_seq2(data_array[:tmax,4*i].tolist(),data_array[:tmax,4*i+1].tolist(),0.7,0.2)
    plot_seq2(data_array[:tmax,4*i+2].tolist(),data_array[:tmax,4*i+3].tolist(),0.7,0.2)

ax.set_xlim([-240, 75])
ax.set_ylim([-110, 110])

# Define the function |z e^(1-z)| - 1
def szego_function(z,alpha):
    
    return np.abs(z/alpha * np.exp(1 + z/alpha)) - 1

# This function plots a rescaled Szegö curve 
def plot_zego(alpha):
    # Create a grid of complex numbers
    re = np.linspace(-235, 70, 1000)
    im = np.linspace(-100, 100, 1000)
    re, im = np.meshgrid(re, im)
    z = re + 1j * im

    # Compute the function on the grid
    szego_values = szego_function(z,alpha)

    ax.contour(re, im, szego_values, levels=[0], colors='blue', linewidths=2)

plot_zego(235)


plt.savefig('Charlier-N=40.pdf')
```
